## Extracción y análisis de HTML

### Paso 1: Realizar una solicitud GET para obtener el HTML de una página

```python
# Realizar una solicitud GET a la página y Leer el código HTML de la página web.
# Make a GET request
req = requests.get('http://www.ilga.gov/senate/default.asp')
# Read the content of the server’s response
src = req.text
# View some output
print(src[:1000])
```

### Paso 2: Analiza la página con Beautiful Soup

```python
# Usamos la función BeautifulSoup para analizar la respuesta en un árbol HTML
# Parse the response into an HTML tree
soup = BeautifulSoup(src, 'lxml')
# Take a look
print(soup.prettify()[:1000])
```

### Paso 3: Buscar elementos HTML

```python
# Beautiful Soup cuenta con varias funciones para encontrar componentes útiles en una página
# Find all elements with a certain tag
a_tags = soup.find_all("a")
print(a_tags[:10])
```

```python
#Lineas equivalentes a soup.find_all("a", href=True) para encontrar todos los enlaces
a_tags = soup.find_all("a")
a_tags_alt = soup("a")
print(a_tags[0])
print(a_tags_alt[0])
```

```python
# Numero de enlaces encontrados
print(len(a_tags))
```

```python
#  En el siguiente codigo, buscamos todas las etiquetas "a" y luego filtramos aquellas con "class_="sidemenu".
# Get only the 'a' tags in 'sidemenu' class
side_menus = soup("a", class_="notranslate")
side_menus[:5]
```

```python
# obtenemos elementos con  "a.sidemenu" CSS Selector.
selected = soup.select("a.notranslate")
selected[:5]
```

### Reto 3: Obtener elementos href que apunten a las facturas de los miembros

```python
def get_members(url):
   # Make a GET request
    req = requests.get(url)
    # Read the content of the server’s response
    src = req.text
    # Soup it
    soup = BeautifulSoup(src, "lxml")
    # Create empty list to store our data
   # Define storage list
    members = []
    members = set()
    # Get rid of junk rows
    valid_rows = [row for row in rows if row.select('div.member-overlay')]
    # Loop through all rows
    for row in valid_rows:
       name = row.select_one('a.notranslate').text.strip()
       #district = example_row.select_one('p.card-text').text.strip()
       print("Name:", name)
       #print("District info:", district)
       p_tag = row.select_one('p.card-text')
       position = p_tag.contents[0].strip()
       print("Party:", position)
       district_info = p_tag.contents[-1].strip()
       print("District info:", district_info)
       #Obtener ref
       a_tag = row.select_one('a.notranslate')
       href = a_tag['href'].strip()  # Obtiene el href
       full_path = "http://www.ilga.gov" + href #+ "&Primary=True"
       print("Full_path:", full_path)
       # Store in a tuple
       senator = (name, district_info, position,full_path)
       # Append to list
       #members.append(senator)
       members.add(senator)
       
    #members = list(members) 
    return(list(members))    
```

```python
url = 'http://www.ilga.gov/senate/default.asp?GA=98'
senate_members = get_members(url)
len(senate_members)
```
